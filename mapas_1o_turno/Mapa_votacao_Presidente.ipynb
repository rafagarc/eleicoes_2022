```python
votacao = votacao.drop(columns=['MUNICÍPIO'])
```

```python
votacao.columns = ['codigo_ibge', 'Município', 'UF', 'AREA_km2', 'geometry', 'codigo_tse',
       'uf', 'Capital', 'Lula', 'Lula_pct',
       'Bolsonaro', 'Bolsonaro_pct', 'Ciro',
       'Ciro_pct', 'Tebet', 'Tebet_pct',
       'Outros', 'Outros_pct', 'Abstencoes',
       'Abstencoes_pct', 'Brancos', 'Brancos_pct',
       'Nulos', 'Nulos_pct', 'BvsL']
```

```python
votacao = votacao[['codigo_ibge', 'codigo_tse', 'Município', 'UF', 'Capital', 'AREA_km2', 
                   'Lula', 'Lula_pct', 'Bolsonaro', 'Bolsonaro_pct',
                   'Ciro', 'Ciro_pct', 'Tebet', 'Tebet_pct',
                   'Outros', 'Outros_pct', 'Abstencoes', 'Abstencoes_pct',
                   'Brancos', 'Brancos_pct', 'Nulos', 'Nulos_pct', 'BvsL', 'geometry']]
```

```python
votacao['Eleitorado'] = votacao[['Lula', 'Bolsonaro', 'Ciro','Tebet', 'Outros', 'Nulos', 'Brancos', 'Abstencoes']].sum(axis=1)
```

```python
votacao.plot(ax=ufs.plot(figsize=(18,18), facecolor='#00000000', edgecolor='#eee', linewidth=2, zorder=2),
             column='Abstencoes_pct', figsize=(18,18), cmap='YlOrBr', edgecolor='face', legend=True,
            legend_kwds={'shrink': 0.8, 'label':'\nAbstenções (%)'})
```

```python
votacao.plot(ax=ufs.plot(figsize=(20,20), facecolor='#00000000', edgecolor='#ff9999', linewidth=2, zorder=2),
             column='Nulos_pct', figsize=(18,18), cmap='Greys', edgecolor='face', legend=True,
            legend_kwds={'shrink': 0.8, 'label':'\nNulos (%)'})
```

```python
votacao.plot(ax=ufs.plot(figsize=(20,20), facecolor='#00000000', edgecolor='#ff9999', linewidth=2, zorder=2),
             column='Brancos_pct', figsize=(18,18), cmap='Greys_r', edgecolor='face', legend=True,
            legend_kwds={'shrink': 0.8, 'label':'\Brancos (%)'})
```

```python
votacao['votos_soltos'] = votacao[['Ciro_pct', 'Tebet_pct', 'Outros_pct']].sum(axis=1).round(2)
```

```python
votacao.plot(ax=ufs.plot(figsize=(20,20), facecolor='#00000000', edgecolor='#dddddd', linewidth=2, zorder=2),
             column='votos_soltos', figsize=(18,18), cmap='inferno', edgecolor='face', legend=True,
            legend_kwds={'shrink': 0.8, 'label':'\nVotação de candidatos que não foram ao 2º turno (%)'})
```
