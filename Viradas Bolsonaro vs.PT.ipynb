```python
nom_2018 = pd.read_csv('votacao_candidato_munzona_2018_BRASIL.csv', sep=';', encoding='ISO-8859-1', decimal=',', thousands='.')
nom_2018 = nom_2018.query('DS_CARGO == "Presidente"')
nom_2018 = nom_2018.query('NR_TURNO == 1')
nom_2018 = nom_2018.groupby(['CD_MUNICIPIO', 'NM_MUNICIPIO', 'SG_UF', 'NM_URNA_CANDIDATO']).QT_VOTOS_NOMINAIS.sum().reset_index()
nom_2018 = nom_2018.pivot(index=['CD_MUNICIPIO', 'NM_MUNICIPIO', 'SG_UF'], columns=['NM_URNA_CANDIDATO'], values='QT_VOTOS_NOMINAIS').fillna(0).astype(int).reset_index()
nom_2018 = nom_2018.query('SG_UF != "ZZ"')
```

```python
det_2018 = pd.read_csv('detalhe_votacao_munzona_2018_BRASIL.csv', sep=';', encoding='ISO-8859-1', decimal=',', low_memory=False, thousands='.')
det_2018 = det_2018.query('DS_CARGO == "Presidente"')
det_2018 = det_2018.query('NR_TURNO == 1')
det_2018 = det_2018.groupby(['CD_MUNICIPIO', 'NM_MUNICIPIO', 'SG_UF'])['QT_VOTOS_NULOS', 'QT_VOTOS_BRANCOS', 'QT_ABSTENCOES'].sum().reset_index()
det_2018 = det_2018.query('SG_UF != "ZZ"')
mrg_nom_2018 = nom_2018[[i for i in nom_2018.columns if i not in ['NM_MUNICIPIO', 'SG_UF']]]
total_2018 = det_2018.merge(mrg_nom_2018, how='left', on='CD_MUNICIPIO')
```

```python
nom_2022 = pd.read_csv('votacao_candidato_munzona_2022_BRASIL.csv', sep=';', encoding='ISO-8859-1', decimal=',', thousands='.')
nom_2022 = nom_2022.query('DS_CARGO == "Presidente"')
nom_2022 = nom_2022.query('NR_TURNO == 1')
nom_2022 = nom_2022.groupby(['CD_MUNICIPIO', 'NM_MUNICIPIO', 'SG_UF', 'NM_URNA_CANDIDATO']).QT_VOTOS_NOMINAIS.sum().reset_index()
nom_2022 = nom_2022.pivot(index=['CD_MUNICIPIO', 'NM_MUNICIPIO', 'SG_UF'], columns=['NM_URNA_CANDIDATO'], values='QT_VOTOS_NOMINAIS').fillna(0).astype(int).reset_index()
nom_2022 = nom_2022.query('SG_UF != "ZZ"')
```

```python
det_2022 = pd.read_csv('detalhe_votacao_munzona_2022_BRASIL.csv', sep=';', encoding='ISO-8859-1', decimal=',', low_memory=False, thousands='.')
det_2022 = det_2022.query('DS_CARGO == "Presidente"')
det_2022 = det_2022.query('NR_TURNO == 1')
det_2022 = det_2022.groupby(['CD_MUNICIPIO', 'NM_MUNICIPIO', 'SG_UF'])['QT_VOTOS_NULOS', 'QT_VOTOS_BRANCOS', 'QT_ABSTENCOES'].sum().reset_index()
det_2022 = det_2022.query('SG_UF != "ZZ"')
mrg_nom_2022 = nom_2022[[i for i in nom_2022.columns if i not in ['NM_MUNICIPIO', 'SG_UF']]]
total_2022 = det_2022.merge(mrg_nom_2022, how='left', on='CD_MUNICIPIO')
```

```python
total_2018['VITORIA_2018'] = total_2018.apply(lambda x: 'BOLSONARO VENCE' if x['JAIR BOLSONARO'] > x['FERNANDO HADDAD'] else ('PT VENCE' if x['JAIR BOLSONARO'] < x['FERNANDO HADDAD'] else 'EMPATE'), axis=1)
```

```python
total_2022['VITORIA_2022'] = total_2022.apply(lambda x: 'BOLSONARO VENCE' if x['JAIR BOLSONARO'] > x['LULA'] else ('PT VENCE' if x['JAIR BOLSONARO'] < x['LULA'] else 'EMPATE'), axis=1)
```

```python
total_2018['MARGEM_VOTOS_2018'] = total_2018['FERNANDO HADDAD']-total_2018['JAIR BOLSONARO']
```

```python
total_2022['MARGEM_VOTOS_2022'] = total_2022['LULA']-total_2022['JAIR BOLSONARO']
```

```python
total_2022['TOTAL_VOTOS'] = total_2022[['QT_VOTOS_NULOS',
       'QT_VOTOS_BRANCOS', 'CIRO GOMES',
       'CONSTITUINTE EYMAEL', 'FELIPE D AVILA', 'JAIR BOLSONARO', 'LULA',
       'LÉO PÉRICLES', 'PADRE KELMON', 'SIMONE TEBET', 'SOFIA MANZANO',
       'SORAYA THRONICKE', 'VERA']].sum(axis=1)
```

```python
total_2018['TOTAL_VOTOS'] = total_2018[['QT_VOTOS_NULOS',
       'QT_VOTOS_BRANCOS', 'ALVARO DIAS', 'CABO DACIOLO',
       'CIRO GOMES', 'EYMAEL', 'FERNANDO HADDAD', 'GERALDO ALCKMIN',
       'GUILHERME BOULOS', 'HENRIQUE MEIRELLES', 'JAIR BOLSONARO',
       'JOÃO AMOÊDO', 'JOÃO GOULART FILHO', 'MARINA SILVA', 'VERA']].sum(axis=1)
```

```python
total_2018['BOLSONARO_PCT_2018'] = (total_2018['JAIR BOLSONARO']/total_2018['TOTAL_VOTOS']*100).round(2)
```

```python
total_2022['BOLSONARO_PCT_2022'] = (total_2022['JAIR BOLSONARO']/total_2022['TOTAL_VOTOS']*100).round(2)
```

```python
total_2018['PT_PCT_2018'] = (total_2018['FERNANDO HADDAD']/total_2018['TOTAL_VOTOS']*100).round(2)
```

```python
total_2022['PT_PCT_2022'] = (total_2022['LULA']/total_2022['TOTAL_VOTOS']*100).round(2)
```

```python
total_2018['MARGEM_PCT_2018'] = (total_2018['PT_PCT_2018']- total_2018['BOLSONARO_PCT_2018']).round(2)
```

```python
total_2022['MARGEM_PCT_2022'] = (total_2022['PT_PCT_2022']- total_2022['BOLSONARO_PCT_2022']).round(2)
```

```python
total_2018 = total_2018.rename(columns={'JAIR BOLSONARO':'JAIR BOLSONARO (2018)'})
```

```python
total_2022 = total_2022.rename(columns={'JAIR BOLSONARO':'JAIR BOLSONARO (2022)'})
```

```python
total_2018 = total_2018.rename(columns={'TOTAL_VOTOS':'TOTAL_VOTOS_2018'})
```

```python
total_2022 = total_2022.rename(columns={'TOTAL_VOTOS':'TOTAL_VOTOS_2022'})
```
