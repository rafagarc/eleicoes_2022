```python
dataset = 'votacao_candidato_munzona_2018_BRASIL.csv'
```

```python
df = pd.read_csv(dataset, sep=';', encoding='ISO-8859-1', decimal=',', thousands='.')
```

```python
#Filtra para ter só dados de governador primeiro turno

dff = df.query('NR_TURNO == 1')
dff = dff.query('DS_CARGO == "Deputado Estadual"')
dff = dff.query('SG_UF == "SP"')
```

```python
dff = dff.groupby(['SQ_CANDIDATO',
       'NM_CANDIDATO', 'NM_URNA_CANDIDATO',
       'DS_SITUACAO_CANDIDATURA', 'DS_DETALHE_SITUACAO_CAND', 'SG_PARTIDO',
       'DS_SIT_TOT_TURNO']).QT_VOTOS_NOMINAIS.sum().reset_index()
```

```python
assembleia = dff[(dff.DS_SIT_TOT_TURNO == 'ELEITO POR QP') | (dff.DS_SIT_TOT_TURNO == 'ELEITO POR MÉDIA')]
```

```python
assembleia.groupby('SG_PARTIDO').SQ_CANDIDATO.count().reset_index().sort_values(by='SQ_CANDIDATO', ascending=False)
```
