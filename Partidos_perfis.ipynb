```python
dataset_2018 = r'C:\Users\rgarcia\Documents\GitHub\eleicoes_2022\consulta_cand_2018_BRASIL.csv'
dataset_2022 = r'C:\Users\rgarcia\Documents\GitHub\eleicoes_2022\consulta_cand_2022_BRASIL.csv'

cand_2018 = pd.read_csv(dataset_2018, sep=';', encoding='ISO-8859-1', decimal=',', low_memory=False)
cand_2022 = pd.read_csv(dataset_2022, sep=';', encoding='ISO-8859-1', decimal=',', low_memory=False)
```

```python
# No dataset de 2018, PL, e CIDADANIA já tinham registro, mas ainda não tinha candidatos.
# Os partidos ainda estava usando os nomes antigos ao mesmo tempo, então isso precisa ser atualizado.

cand_2018['SG_PARTIDO'] = cand_2018['SG_PARTIDO'].replace('PL', 'PR')
cand_2018['SG_PARTIDO'] = cand_2018['SG_PARTIDO'].replace('CIDADANIA', 'PPS')
```

```python
partido = 'PT'
eleicao = '2022'

exec(f"{partido}_{eleicao} = cand_{eleicao}[cand_{eleicao}.SG_PARTIDO == '{partido}'].reset_index(drop=True)")
exec(f"perfil_{partido}_{eleicao} = {partido}_{eleicao}.groupby(['DS_OCUPACAO' ]).SQ_CANDIDATO.count().reset_index().sort_values(by='SQ_CANDIDATO', ascending=False).reset_index( drop=True).reset_index().head(10)")
exec(f"perfil_{partido}_{eleicao}['index'] = perfil_{partido}_{eleicao}['index']+1")
exec(f"perfil_{partido}_{eleicao}.columns = ['Posição', 'Profissão', 'Candidatos']")
exec(f"perfil_{partido}_{eleicao}['Profissão'] = perfil_{partido}_{eleicao}['Profissão'].str.capitalize()")
exec(f"perfil_{partido}_{eleicao}.to_csv('perfil_{partido}_{eleicao}.csv', index=False)")
exec(f"print('    {partido} {eleicao}', '\\n\\n', perfil_{partido}_{eleicao})")

```
