```python
cand = pd.read_csv(r'consulta_cand_2022_BRASIL.csv', sep=';', encoding='ISO-8859-1', decimal=',', thousands='.')

cand = cand.iloc[:, [15, 46, 52]]
```

```python
df = pd.read_csv(r'despesas_contratadas_candidatos_2022_BRASIL.csv', sep=';', encoding='ISO-8859-1', decimal=',', thousands='.')
```

```python
df = df.iloc[:, [9, 12, 15, 17, 18, 20, 24, 48, 52]]
```

```python
df = df.merge(cand, how='left', on='SQ_CANDIDATO')
```

```python
df.info()
```

```python
final = df[df.TP_PRESTACAO_CONTAS == 'Final']
```

```python
final = final.groupby(['SQ_CANDIDATO', 'SG_PARTIDO', 'DS_GENERO', 'DS_COR_RACA']).VR_DESPESA_CONTRATADA.sum().reset_index()
```

```python
genero_partido = final.groupby(['SG_PARTIDO', 'DS_GENERO']).VR_DESPESA_CONTRATADA.sum().reset_index().pivot(index='SG_PARTIDO', columns='DS_GENERO', values='VR_DESPESA_CONTRATADA').fillna(0)
```

```python
genero_partido.to_csv('genero_partido_final.csv')
```

```python
raca_partido = final.groupby(['SG_PARTIDO', 'DS_COR_RACA']).VR_DESPESA_CONTRATADA.sum().reset_index().pivot(index='SG_PARTIDO', columns='DS_COR_RACA', values='VR_DESPESA_CONTRATADA').fillna(0)
```

```python
raca_partido.to_csv('raca_partido_final.csv')
```

```python
genero_partido['COTA'] = genero_partido.FEMININO/(genero_partido.FEMININO+genero_partido.MASCULINO)
```

```python
genero_partido['COTA'] = (genero_partido['COTA']*100).round(1)
```

```python
genero_partido = genero_partido.rename(columns={'COTA':'COTA MULHERES (%)'})
```
