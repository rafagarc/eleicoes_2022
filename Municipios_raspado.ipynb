```python
df = pd.read_csv('candidatos_governador_por_municipio_separado.csv')
```

```python
df = df.query('ESTADO == "SP"')
```

```python
df = df.pivot(index='MUNICíPIO', columns='NOME DO CANDIDATO', values='VOTOS')
```

```python
df['TOTAL'] = df.sum(axis=1)
```

```python
df['VENCEDOR'] = df.drop(columns=['TOTAL']).idxmax(axis=1)
```
