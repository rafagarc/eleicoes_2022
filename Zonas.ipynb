```python
df = pd.read_csv('Lula_vs_Bolsonaro_votacao_SP.csv')
```

```python
df.head()
```

```python
zonas_sp = gpd.read_file('Zonas_Eleitorais_SP.json')
```

```python
zonas_sp = zonas_sp.rename(columns={'ZEFINAL':'Zona'})
```

```python
votacao = zonas_sp.merge(df, how='right', on='Zona')
```

```python
votacao['Vencedor'] = votacao[['Lula (votos)', 'Bolsonaro (votos)', 'Tebet (votos)']].idxmax(axis=1)
```

```python
votacao['Vencedor'] = votacao['Vencedor'].str.replace(' (votos)', '', regex=False)
```

```python
cmap_tab10 = sns.color_palette("tab10", 10)
cmap_new = sns.color_palette([cmap_tab10[0], cmap_tab10[1]], as_cmap=True)
```

```python
votacao.plot(column='Vencedor', categorical=True, cmap='coolwarm', figsize=(20,20), legend=True)
```

```python
votacao = votacao[['FIRST_NOME', 'geometry', 'Lula (%)', 'Lula (votos)', 'Bolsonaro (%)',
       'Bolsonaro (votos)', 'Tebet (%)', 'Tebet (votos)', 'Vencedor']]
```

```python
votacao['FIRST_NOME'] =  votacao.FIRST_NOME.str.title()
```

```python
zonas_sp['FIRST_NOME'] =  zonas_sp.FIRST_NOME.str.title()
```

```python
votacao = votacao.merge(zonas_sp[['Zona', 'FIRST_NOME']], how='left', on='FIRST_NOME')
```

```python
votacao = votacao[['Zona', 'FIRST_NOME', 'Lula (%)', 'Lula (votos)', 'Bolsonaro (%)',
       'Bolsonaro (votos)', 'Tebet (%)', 'Tebet (votos)', 'Vencedor', 'geometry']]
```

```python
votacao = votacao.rename(columns={'FIRST_NOME':'Nome'})
```

```python
votacao.plot(column='Lula (%)', cmap='Reds', figsize=(20,20), legend=True,
            legend_kwds={'label':'\nVotação de Lula (%)'})
```

```python
votacao.plot(column='Bolsonaro (%)', cmap='Blues', figsize=(20,20), legend=True,
            legend_kwds={'label':'\nVotação de Bolsonaro (%)'})
```

```python
votacao.plot(column='Tebet (%)', cmap='Greens', figsize=(20,20), legend=True,
            legend_kwds={'label':'\nVotação de Tebet (%)'})
```

```python
votacao.sort_values(by='Bolsonaro (votos)', ascending=False)
```

```python
votacao.drop(columns=['geometry']).to_csv('votacao_presidente_zona_capital_sp.csv', index=False)
```

```python
votacao.to_file('mapa_votacao_presidente_zona_capital_sp.json', driver='GeoJSON')
```
